```python
import snscrape.modules.twitter as sntwitter
import pandas as pd


# Creating list to append tweet data to
Tweets_dfsept = []
# Using TwitterSearchScraper to scrape data and append tweets
for i,tweetsept in enumerate(sntwitter.TwitterSearchScraper('banjir jakarta since:2022-09-01 until:2022-09-30').get_items()):
    if i>10000:
        break
    Tweets_dfsept.append(tweetsept)

# Creating a dataframe from the tweets list above
Tweets_dfsept2 = pd.DataFrame(Tweets_dfsept)
Tweets_dfsept2
```

```python
# convert date time to date
Tweets_dfsept2['date'] = Tweets_dfsept2['date'].apply(lambda a: pd.to_datetime(a).date()) 
# save to excel
Tweets_dfsept2.to_excel('banjirJakarta2022sept.xlsx')
```
